我是一個開發人員，先幫我建立四個串列student_ids，names、scores和genders，分別存放使用者輸入學生的學號、學生的姓名、學生的成績、學生的性別，幫我撰寫新增、修改、刪除和查詢全部學生的學號、學生的姓名、學生的成績、學生的性別，最後寫一個新增、修改、刪除、查詢全部學生和離開的使用者介面。

````python
# prompt: 我是一個開發人員，先幫我建立四個串列student_ids，names、scores和genders，分別存放使用者輸入學生的學號、學生的姓名、學生的成績、學生的性別，幫我撰寫新增、修改、刪除和查詢全部學生的學號、學生的姓名、學生的成績、學生的性別，最後寫一個新增、修改、刪除、查詢全部學生和離開的使用者介面。

student_ids = []
names = []
scores = []
genders = []

def add_student(student_id, name, score, gender):
  """新增學生資料"""
  student_ids.append(student_id)
  names.append(name)
  scores.append(score)
  genders.append(gender)
  print("新增成功！")

def update_student(student_id, name=None, score=None, gender=None):
  """修改學生資料"""
  try:
    index = student_ids.index(student_id)
    if name is not None:
      names[index] = name
    if score is not None:
      scores[index] = score
    if gender is not None:
      genders[index] = gender
    print("修改成功！")
  except ValueError:
    print("找不到該學號的學生。")

def delete_student(student_id):
  """刪除學生資料"""
  try:
    index = student_ids.index(student_id)
    student_ids.pop(index)
    names.pop(index)
    scores.pop(index)
    genders.pop(index)
    print("刪除成功！")
  except ValueError:
    print("找不到該學號的學生。")

def display_all_students():
  """查詢所有學生資料"""
  if not student_ids:
    print("目前沒有學生資料。")
    return
  print("學號\t姓名\t成績\t性別")
  print("------------------------")
  for i in range(len(student_ids)):
    print(f"{student_ids[i]}\t{names[i]}\t{scores[i]}\t{genders[i]}")

def main():
  """使用者介面"""
  while True:
    print("\n請選擇操作：")
    print("1. 新增學生")
    print("2. 修改學生")
    print("3. 刪除學生")
    print("4. 查詢所有學生")
    print("5. 離開")

    choice = input("請輸入選項：")

    if choice == '1':
      student_id = input("請輸入學號：")
      name = input("請輸入姓名：")
      score = input("請輸入成績：")
      gender = input("請輸入性別：")
      add_student(student_id, name, score, gender)
    elif choice == '2':
      student_id = input("請輸入要修改的學號：")
      print("請輸入要修改的資料 (若不修改請留空)：")
      name = input("姓名：")
      score = input("成績：")
      gender = input("性別：")
      update_student(student_id, name if name else None, score if score else None, gender if gender else None)
    elif choice == '3':
      student_id = input("請輸入要刪除的學號：")
      delete_student(student_id)
    elif choice == '4':
      display_all_students()
    elif choice == '5':
      print("程式結束。")
      break
    else:
      print("無效的選項，請重新輸入。")

if __name__ == "__main__":
  main()
```
````

我是一個開發人員，先幫我建立一個學生類別，該物件包含學生的學號、學生的姓名、學生的成績、學生的性別，用一個list來管理全班學生，幫我撰寫新增、修改、刪除和查詢全部學生的學號、學生的姓名、學生的成績、學生的性別，最後寫一個新增、修改、刪除、查詢全部學生和離開的使用者介面。

````python
# prompt: 我是一個開發人員，先幫我建立一個學生類別，該物件包含學生的學號、學生的姓名、學生的成績、學生的性別，用一個SQL list來管理全班學生，幫我撰寫新增、修改、刪除和查詢全部學生的學號、學生的姓名、學生的成績、學生的性別，最後寫一個新增、修改、刪除、查詢全部學生和離開的使用者介面。

class Student:
    def __init__(self, student_id, name, score, gender):
        self.student_id = student_id
        self.name = name
        self.score = score
        self.gender = gender

students = []

def add_student():
    student_id = input("請輸入學號: ")
    name = input("請輸入姓名: ")
    score = float(input("請輸入成績: "))
    gender = input("請輸入性別: ")
    student = Student(student_id, name, score, gender)
    students.append(student)
    print("學生新增成功!")

def update_student():
    student_id = input("請輸入要修改學生的學號: ")
    for student in students:
        if student.student_id == student_id:
            student.name = input(f"請輸入 {student.name} 的新姓名: ")
            student.score = float(input(f"請輸入 {student.name} 的新成績: "))
            student.gender = input(f"請輸入 {student.name} 的新性別: ")
            print("學生資料修改成功!")
            return
    print("找不到該學號的學生.")

def delete_student():
    student_id = input("請輸入要刪除學生的學號: ")
    for i, student in enumerate(students):
        if student.student_id == student_id:
            del students[i]
            print("學生刪除成功!")
            return
    print("找不到該學號的學生.")

def view_all_students():
    if not students:
        print("目前沒有學生資料.")
        return
    print("學號\t姓名\t成績\t性別")
    for student in students:
        print(f"{student.student_id}\t{student.name}\t{student.score}\t{student.gender}")

def main():
    while True:
        print("\n請選擇操作:")
        print("1. 新增學生")
        print("2. 修改學生")
        print("3. 刪除學生")
        print("4. 查詢全部學生")
        print("5. 離開")

        choice = input("請輸入選項 (1-5): ")

        if choice == '1':
            add_student()
        elif choice == '2':
            update_student()
        elif choice == '3':
            delete_student()
        elif choice == '4':
            view_all_students()
        elif choice == '5':
            print("程式結束.")
            break
        else:
            print("無效的選項，請重新輸入.")

if __name__ == "__main__":
    main()
````
